### Class I and VI Requirements

#### 1777 Proposed Daily Ration by General Washington and Board of General Officers [<sup>1</sup>](#fn1)
- 1 1/4 lb. of beef or 1 lb. pork, or 1 1/4 lb. of salt fish. 
- 1 lb. of flour, or soft bread, or 1 lb. of hard bread. 
- 1/2 gill of rum or whisky per day in lieu of beer. 
- 1/2 pt. of rice, or  1 pt. of Indian meal per week. 
- 3 lbs. of candles to 100 men per week
- 24 lbs. of soft soap or 8 lbs. of hard soap per 100 men per week.
- estimated to cost 3 shillings and 4 pence, exclusive of soap and candles

**Note:** cost estimate seems low but it could be a local spike, missing pages between 1740-1790[<sup>4</sup>](#fn4)

```python
import pandas as pd
import datetime
import numpy as np
from prettytable import PrettyTable
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

def total_food_info(soldier_food, days, pax):
    total_weight = sum(soldier_food.values()) * days * pax
    total_food_type = {k:v*days*pax for (k,v) in soldier_food.items()}
    return {'total_weight':total_weight, 'total_food_type':total_food_type}


start = datetime.datetime(1777, 12, 31)
end = datetime.datetime(1778, 6, 6)
months = "index == '1778-01-01' | index == '1778-02-01' | index == '1778-03-01' | index == '1778-04-01' | index == '1778-05-01' | index == '1778-06-01'"

index = pd.date_range(start, end)
df = pd.DataFrame(index=index)

df['fit for duty'] = np.nan
df['present'] = np.nan

df['fit for duty'].loc['1777-12-31'] = 12544
df['fit for duty'].loc['1778-01-19'] = 9012
df['fit for duty'].loc['1778-01-31'] = 8077
df['fit for duty'].loc['1778-02-14'] = 6804
df['fit for duty'].loc['1778-02-28'] = 6917
df['fit for duty'].loc['1778-03-14'] = 6092
df['fit for duty'].loc['1778-03-21'] = 6250
df['fit for duty'].loc['1778-03-30'] = 5937
df['fit for duty'].loc['1778-04-04'] = 7228
df['fit for duty'].loc['1778-04-11'] = 8652
df['fit for duty'].loc['1778-05-02'] = 11555
df['fit for duty'].loc['1778-05-16'] = 11217
df['fit for duty'].loc['1778-05-30'] = 13842
df['fit for duty'].loc['1778-06-06'] = 13513

df['present'].loc['1777-12-31'] = 15044
df['present'].loc['1778-01-19'] = 13684
df['present'].loc['1778-01-31'] = 13536
df['present'].loc['1778-02-14'] = 11940
df['present'].loc['1778-02-28'] = 12554
df['present'].loc['1778-03-14'] = 11835
df['present'].loc['1778-03-21'] = 11594
df['present'].loc['1778-03-30'] = 10998
df['present'].loc['1778-04-04'] = 11983
df['present'].loc['1778-04-11'] = 12888
df['present'].loc['1778-05-02'] = 15052
df['present'].loc['1778-05-16'] = 14665
df['present'].loc['1778-05-30'] = 17385
df['present'].loc['1778-06-06'] = 16651

df = df.ffill()
df.index=pd.to_datetime(df.index)
df['unfit/injured/ill'] = df['present'] - df['fit for duty']
df['1 DOS Gross Weight Class I,VI'] = df['present'] * 2.166
df['1 DOS gross Calories'] = df['present'] * 2465
df['1 DOS Cost'] = df['present'] * 18.5

fig1, (ax1, ax2) = plt.subplots(1,2, figsize=(15,4))
ax1.set_title("Valley Forge: Fit For Duty Soldiers & Present")
df['fit for duty'].plot(kind='line', ax=ax1, color='tab:green')
df['present'].plot(kind='line', ax=ax1, color='tab:red')
ax1.legend()
ax1.set_ylabel('PAX')
ax2.set_title("Valley Forge: Total Unfit")
df['unfit/injured/ill'].plot(kind='line', ax=ax2)
arive = datetime.datetime(1778, 2, 23)
ax2.set_ylabel('PAX')
ax2.legend()
#ax2.plot(arive, df['unfit/injured/ill'].loc['1778-02-23'], color="black", linewidth=20)
plt.show(fig1)

fig2, (ax3, ax4) = plt.subplots(1,2, figsize=(15,4))
df['Gross Weight Consumed'] = df['1 DOS Gross Weight Class I,VI'].cumsum()
df.query(months)['Gross Weight Consumed'].plot(kind='bar', ax=ax3, color='tab:blue')
ax3.set_title("Gross Weight Consumed Class III & VI")
ax4.set_title("Gross Cost Class III & VI")
#(df.query(months))['1 DOS Gross Weight'].plot(kind='bar', ax=ax3, color='tab:green')

df['Gross Cost'] = df['1 DOS Cost'].cumsum()
df.query(months)['Gross Weight Consumed'].plot(kind='bar', ax=ax4, color='tab:green')

plt.show(fig2)

df[['Gross Cost', 'Gross Weight Consumed']].tail(1)

```

### Class III (AKA Forage) Requirements



#### Thomas Mifflin was reappointed Quartermaster General in 1776, the increased demands for forage in New York required his immediate had indicated were necessary for the support of the main army. Congress furnished him 300,000 dollars to procure the supplies he had indicated were necessary for the support of the main army. These supplies included, among other items, 25,000 bushels of Indian com, 15,000 bushels of oats, 10,000 bushels each of rye meal and spe llS, and 1,800 tons of hay. [<sup>13</sup>](#fn13)

#### To provide forage for the horses, the Cavalry was dispersed to Durham, Connecticut; Lancaster, Pennsylvania; and Winchester, Virginia. Forage had to be provided not only for the Cavalry horses but also for the main army's riding and wagon horses. Biddle estimated the total number at 10,000 [<sup>14</sup>](#fn14)

#### An adult horse at maintenance will consume between 2  2.5% of their bodyweight in feed (hay and grain) each day. For example, a 1,000 pound horse fed a 100% hay diet would consume 25 pounds of hay each day. In the Winter i.e. no grass for grazing. [<sup>15</sup>](#fn15) Hay was 60 to 80 dollars a ton [<sup>15</sup>](#fn16)
